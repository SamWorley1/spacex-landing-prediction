```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd

# NASA Johnson Space Center coordinates
nasa_coordinate = [29.559684888503615, -95.0830971930759]

# Initialize the folium map
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)

# Load the SpaceX launch data (assuming the file is already downloaded and named 'spacex_launch_geo.csv')
spacex_csv_file = 'spacex_launch_geo.csv'
spacex_df = pd.read_csv(spacex_csv_file)

# Select relevant sub-columns: 'Launch Site', 'Lat', 'Long', 'class'
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]

# Add a new column 'marker_color' to spacex_df
def get_marker_color(class_value):
    return 'green' if class_value == 1 else 'red'

spacex_df['marker_color'] = spacex_df['class'].apply(get_marker_color)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(site_map)

# Add markers to the MarkerCluster
for index, row in spacex_df.iterrows():
    folium.Marker(
        location=[row['Lat'], row['Long']],
        popup=f"{row['Launch Site']} (Lat: {row['Lat']}, Long: {row['Long']}, Class: {row['class']})",
        icon=folium.Icon(icon='rocket', color=row['marker_color'])
    ).add_to(marker_cluster)

# Display the map
site_map
```

# TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


You can calculate the distance between two points on the map based on their `Lat` and `Long` values using the following method:


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
!pip install geopy
import folium
from folium.plugins import MousePosition, MarkerCluster
from geopy.distance import geodesic
import pandas as pd

# NASA Johnson Space Center coordinates
nasa_coordinate = [29.559684888503615, -95.0830971930759]

# Initialize the folium map
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)

# Load the SpaceX launch data (assuming the file is already downloaded and named 'spacex_launch_geo.csv')
spacex_csv_file = 'spacex_launch_geo.csv'
spacex_df = pd.read_csv(spacex_csv_file)

# Select relevant sub-columns: 'Launch Site', 'Lat', 'Long', 'class'
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]

# Add a new column 'marker_color' to spacex_df
def get_marker_color(class_value):
    return 'green' if class_value == 1 else 'red'

spacex_df['marker_color'] = spacex_df['class'].apply(get_marker_color)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(site_map)

# Add markers to the MarkerCluster
for index, row in spacex_df.iterrows():
    folium.Marker(
        location=[row['Lat'], row['Long']],
        popup=f"{row['Launch Site']} (Lat: {row['Lat']}, Long: {row['Long']}, Class: {row['class']})",
        icon=folium.Icon(icon='rocket', color=row['marker_color'])
    ).add_to(marker_cluster)

# Add MousePosition to the map
MousePosition().add_to(site_map)

# Define a function to calculate the distance
def calculate_distance(point1, point2):
    return geodesic(point1, point2).kilometers

# Add click event to mark the closest coastline point and calculate distance
def add_coastline_marker(event):
    # Get the coordinates of the click
    coast_point = (event.latlng.lat, event.latlng.lng)
    # Assuming the launch site is the first one for this example
    launch_site = (spacex_df.iloc[0]['Lat'], spacex_df.iloc[0]['Long'])
    # Calculate the distance
    distance = calculate_distance(launch_site, coast_point)
    # Add a marker for the coastline point
    folium.Marker(
        location=[coast_point[0], coast_point[1]],
        popup=f"Coastline Point (Lat: {coast_point[0]}, Long: {coast_point[1]})<br>Distance to Launch Site: {distance:.2f} km",
        icon=folium.Icon(icon='info-sign', color='blue')
    ).add_to(site_map)
    # Re-display the map
    site_map._repr_html_()

# Add the click event to the map
site_map.add_child(folium.ClickForMarker(popup='Click to mark the closest coastline point'))

# Display the map
site_map
```
